```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10, 6))  # Create a figure and axis with specified size

# setting background color
fig.set_facecolor('yellow') 
ax.set_facecolor('lightblue')  # Set the background color of the plot area

# setting title and labels
ax.set_title('Simple Line Chart', fontsize=20, color='blue')  # Set the title with font size and color
ax.set_xlabel('X-axis Label', fontsize=14, color='green')  # Set the x-axis label
ax.set_ylabel('Y-axis Label', fontsize=14, color='green')  # Set the y-axis label

x = [1,2,3,4,5,6,7,8,9] # X-axis data
y = [15,24,56,45,36,26,67,45,10]  # Y-axis data
ax.plot(x,y,marker='o',markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,markersize=8,color='green',linestyle=':')# Plot the data with specific style
ax.axhline(30, color='black',alpha=0.1)
ax.axvline(5, color='black',alpha=0.1)
ax.text(2.5, 33, 'Text Annotation', fontsize=12, color='purple')  # Add text annotation
```
